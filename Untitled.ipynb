```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
mouse_data = "data/Mouse_metadata.csv"
study_data = "data/Study_results.csv"
mouse = pd.read_csv(mouse_data)
study = pd.read_csv(study_data)
```

```python
combined = pd.merge(mouse, study, how="inner", on="Mouse ID")
combined
```

```python
clean = combined.drop_duplicates("Mouse ID")
clean
```

```python
mean = clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
var = clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
s_d = clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
summary = pd.DataFrame({"Mean": mean, "Median": median, "Variance": var, "Standard Deviation": s_d, "SEM": sem})
summary
```

```python
drug = pd.DataFrame(clean.groupby(["Drug Regimen"]).count()).reset_index()
drugs = drug[["Drug Regimen", "Mouse ID"]]
drugs = drugs.set_index("Drug Regimen")
```

```python
drugs.plot(kind="bar", color=['grey'], figsize=(12,5))
plt.title("Drug Treatment Count")
plt.show()
```
